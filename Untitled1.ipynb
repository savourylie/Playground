```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
men_heights = np.random.normal(170, 6, 11000)
women_heights = np.random.normal(163, 6, 10000)
children_heights = np.random.normal(138, 10, 10000)
all_heights = np.concatenate((men_heights, women_heights, children_heights))
```

```python
plt.figure(figsize=(16, 9))
sns.distplot(men_heights, bins=20)
sns.distplot(women_heights, bins=20)
sns.distplot(children_heights, bins=20)
```

```python
plt.figure(figsize=(16, 9))
sns.distplot(all_heights, bins=20)
```
